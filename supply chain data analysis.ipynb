```python
import pandas as pd

# Load dataset
df = pd.read_csv('DataCoSupplyChain.csv', encoding='latin1')

# Select important columns
df = df[[
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Sales',
    'Late_delivery_risk',
    'Order Region'
]]

# ✅ Check size BEFORE cleaning
print("Before Cleaning:", len(df))

# ✅ Remove duplicates
df = df.drop_duplicates()

# ✅ Check size AFTER cleaning
print("After Cleaning:", len(df))

# ✅ Preview again
print("\nPreview after cleaning:\n")
print(df.head())
```

```python
from scipy import stats

# =========================
# HYPOTHESIS TEST 1
# =========================

# Difference between actual and scheduled
df['shipping_diff'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']

# Perform one-sample t-test (testing if mean difference > 0)
t_stat, p_value = stats.ttest_1samp(df['shipping_diff'], 0)

print("\n📊 Hypothesis Test 1 Results")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
# =========================
# HYPOTHESIS TEST 2
# =========================

late_rate = df['Late_delivery_risk'].mean()

print("\nLate Delivery Rate:", round(late_rate * 100, 2), "%")

# One-sample test
t_stat2, p_value2 = stats.ttest_1samp(df['Late_delivery_risk'], 0.4)

print("\n📊 Hypothesis Test 2 Results")
print("t-statistic:", t_stat2)
print("p-value:", p_value2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create visuals folder
if not os.path.exists('visuals'):
    os.makedirs('visuals')

# =========================
# 📊 Visualization 1: Shipping comparison
# =========================

sns.boxplot(data=df[['Days for shipping (real)', 'Days for shipment (scheduled)']])
plt.title("Actual vs Scheduled Shipping Time")
plt.savefig('visuals/shipping_comparison.png')
plt.clf()

# =========================
# 📊 Visualization 2: Late delivery distribution
# =========================

df['Late_delivery_risk'].value_counts().plot(kind='bar')
plt.title("Late vs On-Time Deliveries")
plt.savefig('visuals/late_delivery.png')
plt.clf()

# =========================
# 📊 Visualization 3: Sales by Region (FIXED)
# =========================

plt.figure(figsize=(12, 6))  # ✅ make chart wider

region_sales = df.groupby('Order Region')['Sales'].mean().sort_values()

region_sales.plot(kind='bar')

plt.title("Average Sales by Region")
plt.xticks(rotation=45, ha='right')  # ✅ rotate labels
plt.tight_layout()  # ✅ prevent cutoff

plt.savefig('visuals/region_sales.png')
plt.clf()
```
